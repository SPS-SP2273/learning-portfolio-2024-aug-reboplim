# Exercise 1 (Fill in the blanks)

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='red', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

# Exercise 2 (A simple mathematical plot)

```python
x = np.arange(0, 51, 0.1)
```

```python
y = np.exp(np.negative(x) / 10) * np.sin(x)
```

```python
plt.plot(x, y, color='blue')
plt.xlabel('x', color='darkorange', fontsize=15)
plt.ylabel('y', color='darkorange', fontsize=15, rotation=0)
plt.grid(alpha=.25)
plt.title('$y = e^{-x / 10} sin(x)$', fontsize=20)
plt.ylim(-1, 1)
plt.show()
```
