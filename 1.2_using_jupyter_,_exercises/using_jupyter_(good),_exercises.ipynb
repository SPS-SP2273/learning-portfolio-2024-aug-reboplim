# Exercise 3 (Timing Stuff)

```python
%%timeit
squares = [i ** 2 for i in range(10)]
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i ** 2)
```

- The first value (507 ns) is the mean execution time of the cell.
- The second value (32.4 ns) is the standard deviation of the execution time of the cell.
- The cell is run for 7,000,000 times in total (7 runs and 1,000,000 loops per run).
